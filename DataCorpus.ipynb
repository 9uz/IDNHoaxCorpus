```python
import pandas as pd


dataset= pd.read_csv('/content/drive/MyDrive/Dataset/hoaxs/datasetUMPOHoax.csv')
dataset.head()
```

```python
import matplotlib.pyplot as plt

d=dataset.groupby(['label'])['label'].count()
d.plot.barh(rot=15, title="Jumlah dataset pada setiap label",color = ['b','g','r']);
for index, value in enumerate(d):
    plt.text(value, index,
             str(value))
plt.show(block=True);
```
